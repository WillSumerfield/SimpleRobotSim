```python
import time

from src import ENV_2D, ENV_2_5D
from src.genetic_algorithm import GeneticAlgorithm
from src.agent import CPU_COUNT, train_agent
```

```python
# EA parameters
timesteps_per_generation=3e5
pop_size = 8
generations = 3

def print_stats(start_time, end_time):
    time_diff = end_time - start_time
    print(f"Time taken: {time_diff} seconds, average FPS: {(timesteps_per_generation*pop_size*3)/(time_diff*CPU_COUNT)}")
```

## Test Co-Design FPS

### PyGame Testing

#### Regular RL FPS

```python
start_time = time.time()
train_agent(env_id=ENV_2D, hand_type=0, task_type=0, total_timesteps=timesteps_per_generation*generations*pop_size, verbose=0)
end_time = time.time()
print_stats(start_time, end_time)
```

#### Co-Design FPS

```python
ga = GeneticAlgorithm(env_id=ENV_2D, task_type=0, mutation_rate=0.0, 
                      population_size=pop_size, timesteps_per_generation=timesteps_per_generation)

start_time = time.time()

for generation_idx in range(generations):
        ga.run_generation()
    
end_time = time.time()
print_stats(start_time, end_time)
```
